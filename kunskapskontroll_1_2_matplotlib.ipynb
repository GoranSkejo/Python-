### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Matplotlib allows either procedural or object-oriented (OO) ways to plot. The OO-way is recommended sinse Python is an object-oriented language. OO-way is also better for generating figures with different subplots.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is the foundation on which data is plotted. A figure is automatically generated during plotting. Axes is the plotting area into which most of the objects go. Axes has different helper methods. Axis is used to mark measurements. Artists are everything that is shown on a figure.

### When plotting in Matplotlib, what is the expected input data type?

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

# pyplot approach
plt.plot(x, y)
plt.title("y = x**2")
plt.xlabel('x')
plt.ylabel('y')

# object-oriented approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('y = x**2')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(1, 2,layout='constrained')
fig.suptitle('Two Plots', size = 20)

axs[0].bar(names, values)
axs[0].set_title('Bar Plot')

axs[1].scatter(random_data_x, random_data_y)
axs[1].set_title('Scatter Plot')
```
